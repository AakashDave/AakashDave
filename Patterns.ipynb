```python
# 1. Pattern - 1 https://takeuforward.org/pattern/pattern-1-rectangular-star-pattern/
n = int(input("Enter a number here:"))
for i in range(n):
  print("*"*n)

print("---------------------------------")

for i in range(n):
  for j in range(n):
    print("*", end="")
  print()
```

```python
# Pattern 2. https://takeuforward.org/pattern/pattern-2-right-angled-triangle-pattern/
n = int(input("Enter a number here: "))
for i in range(n):
  print("*" * i)

print("================")

for i in range(n):
  for j in range(i):
    print("*", end="")
  print()
```

```python
# Pattern 3. https://takeuforward.org/pattern/pattern-3-right-angled-number-pyramid/
n = int(input("Enter a number here: "))
for i in range(n):
  for j in range(1, i+1):
    print(j, end="")
  print()
```

```python
# Pattern 4. https://takeuforward.org/pattern/pattern-4-right-angled-number-pyramid-ii/
n = int(input("Enter a number here: "))
for i in range(1, n+1):
  for j in range(i):
    print(i, end="")
  print()
```

```python
# Pattern 5. https://takeuforward.org/pattern/pattern-5-inverted-right-pyramid/
n = int(input("Enter a number here: "))

for i in range(n):
  print("*"*(n-i))

print("==============================")

for i in range(n):
  for j in range(n-i):
    print("*", end="")
  print()
```

```python
# Pattern 6. https://takeuforward.org/pattern/pattern-6-inverted-numbered-right-pyramid/
n = int(input("Enter a number here: "))
for i in range(n):
  for j in range(1, n-i+1):
    print(j, end="")
  print()
```

```python
# Pattern 7. https://takeuforward.org/pattern/pattern-7-star-pyramid/
n = int(input("Enter a number here: "))
for i in range(1, n+1):
  for j in range(n-i):
    print("", end=" ")
  for k in range(2*i - 1):
    print("*", end="")

  print()

print("======================================")

for i in range(1, n + 1):
    spaces = " " * (n - i)
    stars = "*" * (2 * i - 1)
    print(spaces + stars)
```

```python
# Pattern 8. https://takeuforward.org/pattern/pattern-8-inverted-star-pyramid/
n = int(input("Enter a number: "))
for i in range(n):
  spaces = " " * i
  stars = "*" * ((n-i)*2 - 1)
  print(spaces + stars)
```

```python
# Pattern 9. https://takeuforward.org/pattern/pattern-9-diamond-star-pattern/
n = int(input("Enter a number: "))

# Upper half
for i in range(1, n + 1):
    spaces = " " * (n - i)
    stars = "*" * (2 * i - 1)
    print(spaces + stars)

# Lower half
for i in range(n - 1, 0, -1):
    spaces = " " * (n - i)
    stars = "*" * (2 * i - 1)
    print(spaces + stars)

print("===================================")

for i in list(range(1, n+1)) + list(range(n-1, 0, -1)):
  print(" " * (n - i) + "*" * (2 * i - 1))
```

```python
# Pattern 10. https://takeuforward.org/pattern/pattern-10-half-diamond-star-pattern/
n = int(input("Enter a number: "))

for i in list(range(1, n+1)) + list(range(n-1, 0, -1)):
  print("*" * i)
```
